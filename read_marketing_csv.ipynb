### Using multiple conditions to create new features.

```python
df.loc[(df['NumWebPurchases'] > 5) | (df['NumDealsPurchases'] > 5), 'ActiveOnlineBuyers'] = "Active"
df.loc[(df['NumWebPurchases'] < 5) & (df['NumDealsPurchases'] < 5), 'ActiveOnlineBuyers'] = "Not Active"
```

### Campaign for targeting highly educated folks above a certain income with no(zero) or one kid.

```python
df['Education'].value_counts()
```

```python
education_class = ['Graduation', 'PhD', 'Master']
education_criteria = df['Education'].isin(education_class)
salary_criteria = df['Income'] > 65000
kids_criteria = df['Kidhome'] <= 1

final_criteria = (education_criteria & salary_criteria & kids_criteria)
```
